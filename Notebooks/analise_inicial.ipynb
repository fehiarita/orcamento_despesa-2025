```python
#tirar os valores duplicados do csv
ler_csv = ler_csv.drop_duplicates()
ler_csv.head(900)
# Salva o csv 'ler_csv' em um novo arquivo
ler_csv.to_csv('../dados/bruto/2025_OrcamentoDespesa_semduplicadas.csv', index=False, sep=';', encoding='latin-1')
```

```python
#vizualizar os valores duplicados do csv (duplicado não significa que está errado, pode ser que haja mais de uma linha com os mesmos dados)
print(ler_csv[ler_csv.duplicated(keep=False)])
```

```python
# Retorna todas as linhas que são cópias de uma anterior nos seus indexes, ou seja, todas as linhas duplicadas, incluindo a primeira ocorrência.
duplicados_todos = ler_csv[ler_csv.duplicated(keep=False)]
print(duplicados_todos)
```

```python
#tirar valores nulos do csv
ler_csv = ler_csv.dropna()
ler_csv.head(900)
```

```python
#loop criado para padronizar as colunas de orçamento de tipo string para o tipo float
colunas_dinheiro = ['ORÇAMENTO INICIAL (R$)', 'ORÇAMENTO ATUALIZADO (R$)', 'ORÇAMENTO EMPENHADO (R$)', 'ORÇAMENTO REALIZADO (R$)']

for coluna in colunas_dinheiro:
    # Converte para string e usamos o .str para aplicar a substituição de caracteres corretamente
    ler_csv[coluna] = ler_csv[coluna].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
    
    # Converte para float ignorando erros de textos inválidos (viram NaN)
    ler_csv[coluna] = pd.to_numeric(ler_csv[coluna], errors='coerce')
    print(ler_csv[colunas_dinheiro].head())
ler_csv.to_csv(r'C:\Users\nanda\OneDrive\Documentos\estudo\analise de dados\orcamento_despesa-2025\dados\bruto\2025_OrcamentoDespesa_semduplicadas.csv', index=False, sep=';', encoding='utf-8-sig')
```

```python
#loop criado para padronizar as colunas de orçamento de tipo string para ser tudo maiúsculo
colunas_string = ['NOME ÓRGÃO SUPERIOR', 'NOME ÓRGÃO SUBORDINADO', 'NOME UNIDADE ORÇAMENTÁRIA','NOME FUNÇÃO','NOME SUBFUNÇÃO', 'NOME PROGRAMA ORÇAMENTÁRIO', 'NOME AÇÃO', 'NOME CATEGORIA ECONÔMICA', 'NOME GRUPO DE DESPESA','NOME ELEMENTO DE DESPESA']
for coluna in range(len(colunas_string)):
    ler_csv[colunas_string[coluna]] = ler_csv[colunas_string[coluna]].astype(str).str.upper()
    print(ler_csv[colunas_string[coluna]].head())
#usando a formatação utf-8-sig para evitar problemas ao abrir o arquivo csv em outros programas, como o Excel, que podem não reconhecer corretamente os caracteres acentuados.
ler_csv.to_csv(r'C:\Users\nanda\OneDrive\Documentos\estudo\analise de dados\orcamento_despesa-2025\dados\bruto\2025_OrcamentoDespesa_semduplicadas.csv', index=False, sep=';', encoding='utf-8-sig')
```

Análises de dados do csv 

```python
#mostra os 10 orgãos que mais gastaram baseado no orçamento realizado
top10_gastos = ler_csv.groupby('NOME ÓRGÃO SUPERIOR')['ORÇAMENTO REALIZADO (R$)'].sum().sort_values(ascending=False)
pd.set_option('display.float_format', '{:,.2f}'.format)
print(top10_gastos.head(10))
```

```python
agrupamento_por_ministerio = ler_csv.groupby('NOME ÓRGÃO SUPERIOR')['ORÇAMENTO REALIZADO (R$)'].sum().sort_values(ascending=False)
pd.set_option('display.float_format', '{:,.2f}'.format)  # Formata os números para exibir com vírgula como separador de milhares e ponto como separador decimal
print(agrupamento_por_ministerio)
```

Gráficos feitos a partir do csv

```python
import matplotlib.pyplot as plt
import seaborn as sns  

plt.figure(figsize=(16, 6))

top10_gastos = top10_gastos.head(10)

sns.barplot(x=top10_gastos.values, y=top10_gastos.index, palette='Blues_d', edgecolor='black')

plt.title('Top 10 Órgãos que mais gastam por Orçamento Realizado', fontsize=14, pad=15)
plt.xlabel('Orçamento Realizado (R$)', fontsize=12)
plt.ylabel('Órgão Superior', fontsize=12)

plt.ticklabel_format(style='plain', axis='x')
plt.xticks(rotation=45)

plt.tight_layout()

# 1. SALVA PRIMEIRO: Guarda o arquivo na sua pasta antes de limpar a tela
plt.savefig('top10_gastos.png', dpi=300)

# 2. MOSTRA DEPOIS: Exibe o gráfico na tela do seu Jupyter Notebook
plt.show()


```

```python
# 1. Filtra os ministérios principais (buscando pelo nome na coluna)
gasto_saude = ler_csv[ler_csv['NOME ÓRGÃO SUPERIOR'].str.contains('SAÚDE', na=False)]['ORÇAMENTO REALIZADO (R$)'].sum()
gasto_educacao = ler_csv[ler_csv['NOME ÓRGÃO SUPERIOR'].str.contains('EDUCAÇÃO', na=False)]['ORÇAMENTO REALIZADO (R$)'].sum()
gasto_presidencia = ler_csv[ler_csv['NOME ÓRGÃO SUPERIOR'].str.contains('PRESIDÊNCIA DA REPÚBLICA', na=False)]['ORÇAMENTO REALIZADO (R$)'].sum()

# 2. Filtra os gastos da COP 30 (buscando o termo em QUALQUER coluna de texto do arquivo)
# Varremos todas as colunas de texto procurando pela menção "COP 30" ou "COP30"
filtro_cop30 = ler_csv.astype(str).apply(lambda x: x.str.contains('COP 30|COP30', case=False, na=False)).any(axis=1)
gasto_cop30 = ler_csv[filtro_cop30]['ORÇAMENTO REALIZADO (R$)'].sum()

# 3. Cria um novo grupo (Series) organizado com esses resultados
gastos_selecionados = pd.Series({
    'Ministério da Saúde': gasto_saude,
    'Ministério da Educação': gasto_educacao,
    'Presidência da República': gasto_presidencia,
    'Ações da COP 30': gasto_cop30
}).sort_values(ascending=False)

print(gastos_selecionados.map('{:.2f}'.format))

import matplotlib.pyplot as plt
import seaborn as sns

# Define o tamanho ideal para o gráfico horizontal
plt.figure(figsize=(16, 6))

# Cria o gráfico de barras horizontais utilizando uma paleta de cores destacada ('Set2' ou 'viridis')
sns.barplot(x=gastos_selecionados.values, y=gastos_selecionados.index, palette='viridis', edgecolor='black')

plt.title('Comparativo de Despesas: Saúde vs Educação vs Presidência vs COP 30 (2025)', fontsize=14, pad=15)
plt.xlabel('Orçamento Realizado (R$)', fontsize=12)
plt.ylabel('Destino do Gasto', fontsize=12)

# Remove a notação científica e rotaciona os números se necessário
plt.ticklabel_format(style='plain', axis='x')
plt.xticks(rotation=45)

# Adiciona linhas de grade verticais discretas para ajudar na leitura dos valores gigantes
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()

# Salva a imagem para o seu GitHub
plt.savefig('comparativo_gastos_especificos.png', dpi=300)
plt.show()

```
